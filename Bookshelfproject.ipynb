```python
df = df.dropna(axis=1, how='all')
df=df.drop(["Book Id","ISBN"],axis=1)
```

```python
df.dtypes
```

```python
import pandas as pd

# -----------------------------
# 1. Convert ID-like columns to string
# -----------------------------
id_cols = ['ISBN', 'Book Id']
for col in id_cols:
    if col in df.columns:
        df[col] = df[col].astype(str)

# -----------------------------
# 2. Convert date columns to datetime
# -----------------------------
date_cols = [
    'Loan Start Date', 'Loan Due Date',
    'Purchase Date', 'Transfer Date'
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# -----------------------------
# 3. Convert true numeric columns properly
# -----------------------------
true_numeric = [
    'Pages Read', 'Page Count', 'My Rating', 'Volume',
    'Purchase Price', 'Transfer Price'
]

for col in true_numeric:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# -----------------------------
# 4. Fill NaN values in numeric columns
# -----------------------------
fill_rules = {
    'Page Count': 0,
    'Pages Read': 0,
    'Volume': 1,
    'My Rating': df['My Rating'].mean() if 'My Rating' in df.columns else None,
    'Purchase Price': 0,
    'Transfer Price': 0
}

for col, value in fill_rules.items():
    if col in df.columns:
        df[col] = df[col].fillna(value)

# -----------------------------
# 5. Recalculate numeric columns correctly
# -----------------------------
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Final numeric columns:", list(num_cols))
print("Count:", len(num_cols))
```

```python
# Identify numerical columns
cols = df.select_dtypes(include=["int64","float64"]).columns
print("Numerical columns:", list(num_cols))
print("Count:", len(num_cols))
```

```python
df[cols].isna().sum()
```

```python
# Identify Categorical Columns
cat_cols = df.select_dtypes(include=["object"]).columns
print("Categorical columns:", list(cat_cols))
print("Count:", len(cat_cols))
```

```python
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=False)
```

```python
df_encoded.head()
#df_encoded.shape
```

```python
# Identify boolean columns##
bool_cols = df.select_dtypes(include=["bool"]).columns
print("Boolean columns:", list(bool_cols))
print("Count:", len(bool_cols))
```

```python
# Identify date columns
date_cols = df.select_dtypes(include=["datetime64[ns]"]).columns
print("Date columns:", list(date_cols))
print("Count:", len(date_cols))
```
